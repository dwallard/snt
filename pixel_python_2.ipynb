On charge la biblithèque PIL:

```python
from PIL import Image
```

On ouvre la fichier pomme.jpg

```python
img = Image.open('pomme.jpg')
```

Pour afficher les images dans le Notebook, nous avons besoin d'ajouter les lignes suivantes. Ne pas les copier dans votre programme pomme.py (elles ne sont pas nécessaires) mais excécuter les pour la suite du Notebook.

```python
from matplotlib.pyplot import imshow
import numpy as np
```

```python
imshow(np.asarray(img))
```

```python
largeur,hauteur = img.size    # On récupère la définition de l'image
for y in range(hauteur):          # y parcours toute la hauteur de l’image
    for x in range(largeur):      # x parcours toute la largeur de l’image
        r,v,b=img.getpixel((x,y))   # On récupère le codage rvb du pixel
        img.putpixel((x,y),(v,b,r))   # On modifie le codage rvb de même pixel

imshow(np.asarray(img))
```

Recopier ci-dessous et modifier le codage rvb afin d'obtenir une autre image de la pomme.
