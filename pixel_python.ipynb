```python
from PIL import Image
```

Pour afficher les images dans le Notebook, nous avons besoin d'ajouter les lignes suivantes.
Ne pas les copier dans votre programme pomme.py (elles ne sont pas nécessaires) mais excécuter les pour la suite du Notebook.

```python
from matplotlib.pyplot import imshow
import numpy as np
```

On ouvre le fichier image avec l'instruction:

```python
img = Image.open('pomme.jpg')
```

On peut afficher la définition de l'image:
(Ecrire print(img.size) dans votre programme)

```python
img.size
```

Nous allons créer un pixel bleu au milieu de l'image.

```python
bleu = (0,0,255)
img.putpixel((250,250),bleu)
```

Pour afficher le résultat, ajouter l'instruction img.show()
Pour le Notebook, exécuter la ligne suivante:

```python
imshow(np.asarray(img))
```

En zoomant au centre l'image vous devriez voir le pixel.

Sur le Notebook, on ne le voit pas car il est trop petit.
On va donc ajouter 16 pixels bleus.

```python
img.putpixel((250,251),bleu)
img.putpixel((250,252),bleu)
img.putpixel((250,253),bleu)
img.putpixel((251,250),bleu)
img.putpixel((251,251),bleu)
img.putpixel((251,252),bleu)
img.putpixel((251,253),bleu)
img.putpixel((252,250),bleu)
img.putpixel((252,251),bleu)
img.putpixel((252,252),bleu)
img.putpixel((252,253),bleu)
img.putpixel((250,250),bleu)
img.putpixel((253,251),bleu)
img.putpixel((253,252),bleu)
img.putpixel((253,253),bleu)
imshow(np.asarray(img))
```

On peut donc voir au centre un pixel bleu.

Remarque: Pour mettre les 16 pixels en bleu, on peut faire deux boucles comme suit:

```python
for x in range(250,254):
    for y in range(250,254):
        img.putpixel((x,y),bleu)
imshow(np.asarray(img))
```
